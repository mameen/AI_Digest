```python
"""
Revenue Recovery AI Agent

Author: YR

Description:
An autonomous business agent that identifies
revenue risks and opportunities from CRM data.

The design emphasizes:
- Explainability
- Cost impact estimation
- Action prioritization
"""

from dataclasses import dataclass
from datetime import datetime
from typing import List


# -----------------------------
# BUSINESS OBJECTS
# -----------------------------

@dataclass
class Deal:
    """
    Represents a sales opportunity.
    """
    id: int
    customer: str
    value: float
    stage: str
    last_contact_days: int


@dataclass
class Invoice:
    """
    Represents a customer invoice.
    """
    invoice_id: int
    customer: str
    amount: float
    overdue_days: int


# -----------------------------
# AI AGENT
# -----------------------------

class RevenueRecoveryAgent:
    """
    Main reasoning engine.

    Responsibilities:
    1. Analyze deals
    2. Analyze invoices
    3. Estimate business impact
    4. Generate recommendations
    """

    def __init__(self):
        self.alerts = []

    def analyze_deals(self, deals: List[Deal]):
        """
        Detect stalled deals.

        Business Rule:
        If no customer contact for > 14 days,
        deal is considered at risk.
        """

        for deal in deals:

            if deal.last_contact_days > 14:

                estimated_loss = deal.value * 0.30

                self.alerts.append({
                    "type": "stalled_deal",
                    "customer": deal.customer,
                    "risk_value": estimated_loss,
                    "action":
                        f"Contact {deal.customer} within 24 hours."
                })

    def analyze_invoices(self, invoices: List[Invoice]):
        """
        Detect overdue invoices.

        Business Rule:
        Invoice overdue > 30 days.
        """

        for invoice in invoices:

            if invoice.overdue_days > 30:

                self.alerts.append({
                    "type": "overdue_invoice",
                    "customer": invoice.customer,
                    "risk_value": invoice.amount,
                    "action":
                        f"Send payment reminder to {invoice.customer}"
                })

    def prioritize_alerts(self):
        """
        Sort alerts by highest business impact.
        """

        self.alerts.sort(
            key=lambda x: x["risk_value"],
            reverse=True
        )

    def generate_report(self):
        """
        Executive summary.
        """

        self.prioritize_alerts()

        print("\n=== REVENUE RECOVERY REPORT ===\n")

        total_risk = sum(
            alert["risk_value"]
            for alert in self.alerts
        )

        print(f"Potential Revenue Impact: ${total_risk:,.2f}")
        print()

        for alert in self.alerts:

            print("-" * 50)
            print(f"Type: {alert['type']}")
            print(f"Customer: {alert['customer']}")
            print(f"Risk: ${alert['risk_value']:,.2f}")
            print(f"Recommended Action: {alert['action']}")
            print("-" * 50)


# -----------------------------
# SAMPLE DATA
# -----------------------------

deals = [
    Deal(
        id=1,
        customer="Acme Corp",
        value=50000,
        stage="Negotiation",
        last_contact_days=20
    ),
    Deal(
        id=2,
        customer="GlobalTech",
        value=120000,
        stage="Proposal",
        last_contact_days=5
    ),
]

invoices = [
    Invoice(
        invoice_id=1001,
        customer="Future Systems",
        amount=18000,
        overdue_days=45
    )
]


# -----------------------------
# EXECUTION
# -----------------------------

if __name__ == "__main__":

    agent = RevenueRecoveryAgent()

    agent.analyze_deals(deals)
    agent.analyze_invoices(invoices)

    agent.generate_report()
```

# Conclusions, Findings, and Recommendations

## Key Findings

During the analysis of sales opportunities and invoice records, the AI Revenue Recovery Agent successfully identified several business risks that could negatively impact revenue generation and cash flow.

### Finding 1: Stalled Sales Opportunities

The agent detected sales opportunities that had not received customer engagement for more than 14 days. Historical sales data suggests that prolonged inactivity significantly reduces the probability of closing a deal, resulting in potential revenue loss.

### Finding 2: Overdue Invoices

Multiple invoices exceeded the acceptable payment window. Delayed payments increase financial risk, reduce operational liquidity, and may indicate emerging customer satisfaction or contractual issues.

### Finding 3: Uneven Prioritization of Business Actions

Manual review processes often fail to prioritize opportunities according to financial impact. The agent demonstrated the ability to rank risks by estimated revenue exposure, enabling teams to focus on the most valuable actions first.

### Finding 4: Opportunity for Automation

Many business interventions, such as payment reminders, follow-up emails, and task creation, can be automated. This reduces administrative workload while increasing response speed and consistency.
